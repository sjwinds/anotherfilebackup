## Convert to Numpy Array

```python
from PIL import Image
from numpy import asarray

#load the image
image = Image.open("Huaweiswans.JPG")

#convert image to numpy arry
data = asarray(image)

print(type(data))
print(data.size)

#create Pillow image
image2 = Image.fromarray(data)
print(type(image2))

# summarize image details
print(image2.mode)
print(image2.size)
```

# Resizie, Convert images to Grayscale, crop and save as JPG file

```python
import numpy as np

# the size of the original image
width, height = image.size

# resize 1/3 size of the original
width = int(width/3)
height = int(height/3)

# resize the image and crop 
image_1_over_3 = image.resize((width,height))
image_crop = image_1_over_3.crop((0,20, 900,600))

# convert the resized, corped and rotated image to grayscale and save it as 'huiweiswans_g_r.JPG'
image_rr_grayscale = np.array(image_crop.rotate(270).convert('L'))
image_grayscaled = Image.fromarray(image_rr_grayscale).save('huiweiswasn_g_r.JPG')
```
